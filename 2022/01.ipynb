# Day 1 - Calorie Counting
## Data

```python
example_elves = [
    [1000, 2000, 3000],
    [4000],
    [5000, 6000],
    [7000, 8000, 9000],
    [10000]
]

example_elves
```

```python
from aocd import get_data

raw_data = get_data(year=2022, day=1)

elves = []
elf = []

for line in raw_data.split('\n'):
    if len(line) > 0:
        elf.append(int(line))
    else:
        elves.append(elf)
        elf = []
        
elves
```

## Part 1

```python
def get_most_calories(elves):
    elf_total_calories = [sum(elf) for elf in elves]
    
    elf_highest_calories = 0
    
    for calories in elf_total_calories:
        if calories > elf_highest_calories:
            elf_highest_calories = calories
            
    return elf_highest_calories
        
display(get_most_calories(example_elves))
display(get_most_calories(elves))
```

## Part 2

```python
def elf_sort(elves):
    """Return elves sorted by total calories, greatest to least"""
    sorted_elves = []
    
    for elf in elves:
        calories = sum(elf)
        inserted = False
        
        for i in range(len(sorted_elves)):
            
            if calories > sorted_elves[i]:
                sorted_elves.insert(i, calories)
                inserted = True
                break
                
        if inserted == False:
            sorted_elves.append(calories)
                
    return sorted_elves
    
elf_sort(example_elves)
```

```python
def get_top_calories(elves, n):
    """Return the sum of the calories carried by the top N elves."""
    sorted_elves = elf_sort(elves)
    elf_highest_calories = []
    
    if n > len(elves):
        n = len(elves)
        
    for i in range (n):
        elf_highest_calories.append(sorted_elves[i])
    return sum(elf_highest_calories)


display(get_top_calories(example_elves, 3))
display(get_top_calories(elves, 3))
```

```python
def sum_top_n(elves, n):
    total_calories = [sum(elf) for elf in elves]
    sorted_calories = sorted(total_calories)
    return sum(sorted_calories[-n:])

sum_top_n(elves, 3)
```

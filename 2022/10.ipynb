## Parsing

```python

def parse(data):
    """
    Return a list of values representing how the X register is changed by each cycle.
    addx V ==> [0, V]
    noop ==> [0]
    """
    lines = data.split('\n')
    parsed_lines = []
    
    for line in lines:
        parts = line.split(' ')


        parsed_lines.append(0)
        if len(parts) == 2:
            parsed_lines.append(int(parts[1]))

    return parsed_lines

example_x_additions = parse(example_data)
real_x_additions = parse(raw_data)
```

## Part 1

```python
def calc_x_val(x_additions):
    
    x_val = 1
    x_vals = [1]
    
    for step in x_additions:
        x_val += step
        x_vals.append(x_val)
        
    return x_vals

assert calc_x_val(example_x_additions)[19] == 21
assert calc_x_val(example_x_additions)[59] == 19
assert calc_x_val(example_x_additions)[219] == 18

def calc_signal_strength(x_additions):
    """
    Given a sequence of X register additions, return the signal strength where
    signal strength is the sum of the x values multipled by the cycle number at
    the 20th, 60th, 100th, ... cycle.
    """
    
    signal_strength = 0
    
    x_vals = calc_x_val(x_additions)
    
    for cycle in range(20, 221, 40):
        signal_strength += x_vals[cycle - 1] * cycle
    
    return signal_strength
    


assert calc_signal_strength(example_x_additions) == 13140
calc_signal_strength(real_x_additions)
```

## Part 2

```python
def render_screen(x_additions):
    """
    Given a value of x_additions, render the CRT screen.
    Remember a pixel a (row, col) is ON if the x value at cycle = (row * 40) + col is within 1
    of the column.
    """
    num_rows = 6
    num_cols = 40
    x_vals = calc_x_val(x_additions)
    screen = [[' '] * num_cols for row in range(num_rows)]
    
    for row in range(num_rows):
        for col in range(num_cols):
            cycle = (row * num_cols) + col
            x_pos = x_vals[cycle]
            distance = abs(x_pos - col)
            
            if distance <= 1:
                #print(f"row: {row}, col: {col}, cyc: {cycle}, x: {x_pos}, distance; {distance}")
                screen[row][col] = '#'
        
   
    print("0123456789" * 4)
    for row in screen:
        print(''.join(row))
    print("0123456789" * 4)
    

print("Example Screen:")
render_screen(example_x_additions)

print("\n\nReal Screen:")
render_screen(real_x_additions)
```

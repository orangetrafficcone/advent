```python
def fuel_usage_part1(p0, p1):    
    """
    return the amount of fuel used to get from 'p0' to 'p1'
    """
    return abs(p0 - p1)

print(fuel_usage_part1(8, 12))
print(fuel_usage_part1(16, 8))
```

```python
example = [16,1,2,0,4,2,7,1,2,14]

def total_fuel_usage(point, dests, calc_fuel_usage):
    
    """
    return the total fuel usage from 'point' to a list of destinations
    """
    
    total_usage = 0
    for i in range(len(dests)):
        total_usage += calc_fuel_usage(point, dests[i])
    return total_usage

print(total_fuel_usage(2, example, fuel_usage_part1))
```

```python
from matplotlib import pyplot as plt

xs = list(range(-10, 20))
print(xs)
ys = [total_fuel_usage(x, example, fuel_usage_part1) for x in xs]

plt.plot(xs, ys);
```

```python
def find_minimal_fuel_usage(dests, calc_fuel_usage):
    """
    return the mininmal fuel usage for all crab-subs located at 'dests' to align on one point.
    """
    minimal_fuel = None
    for i in range(min(dests), max(dests) + 1):
        fuel = total_fuel_usage(i, dests, calc_fuel_usage)
        if minimal_fuel is None or fuel < minimal_fuel:
            minimal_fuel = fuel
    return minimal_fuel
            
print(find_minimal_fuel_usage(example, fuel_usage_part1))
```

```python
cache = {
    1: 1,
    2: 3,
    3: 6,
    # ...
}

def triangle_number(n):
    """Returns the nth triangle number."""
    if n not in cache:
        cache[n] = sum(range(n+1))
        
    return cache[n]

def fuel_usage_part2(p0, p1):
    """
    Calculates the fuel usage from p0 to p1, given fuel costs are triangular.
    """
    distance = abs(p0 - p1)
    return triangle_number(distance)

print(find_minimal_fuel_usage(example, fuel_usage_part2))
print(cache)
```

```python
# dad's version

def dads_version(crabs, calc_fuel_usage):
    return min([sum(calc_fuel_usage(p, crab) for crab in crabs) for p in list(range(min(crabs), max(crabs)+1))])

print(dads_version(example, fuel_usage_part2))
```

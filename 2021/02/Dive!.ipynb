```python
with open('input.txt', 'r') as f:
    lines = f.read().splitlines() 
    
print(f'Some data: {lines[0:10]}')
print(f'Number of lines: {len(lines)}')
```

# Part 1

```python
depth = 0
horizontal_position = 0
for line in lines:
    split = line.split(' ')
    print(f'{split}')
    if split[0]=='forward':
        horizontal_position += int(split[1])

    if split[0]=='down':
        depth += int(split[1])
    if split[0]=='up':
        depth -= int(split[1])
print(depth)
print(horizontal_position)
print(depth * horizontal_position)
    
```

# Part 2

```python
depth = 0
horizontal_position = 0
aim = 0
xs = [0]
ys = [0]

for line in lines:
    split = line.split(' ')
    print(f'{split}')
    #forward
    if split[0] == 'forward':
        horizontal_position += int(split[1])
        depth += int(split[1]) * aim
    #down
    if split[0] == 'down':
        aim += int(split[1])
    #up
    if split[0] == 'up':
        aim -= int(split[1])
        
    xs.append(horizontal_position)
    ys.append(depth)

print(depth)
print(horizontal_position)
print(depth * horizontal_position)
    
```

```python
import matplotlib.pyplot as plt
plt.gca().invert_yaxis()
plt.ylabel('depth')
plt.plot(xs, ys);
```
